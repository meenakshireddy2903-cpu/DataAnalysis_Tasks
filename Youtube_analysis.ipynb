```python
# Import necessary libraries for data manipulation and visualization.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from a URL into a pandas DataFrame and display its columns.
df = pd.read_csv("https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
df.columns
```

```python
# Display a concise summary of the DataFrame, including data types and non-null values.
df.info()
```

```python
# Check for the count of null (missing) values in each column of the DataFrame.
df.isnull().sum()
```

**#Handling Null Values**

```python
# Display all unique values present in the 'Ranking' column.
df["Ranking"].unique()
```

```python
# Calculate and display the mean value of the 'Ranking' column.
df["Ranking"].mean()
```

```python
# Fill any missing (NaN) values in the 'Ranking' column with its calculated mean.
df["Ranking"] = df["Ranking"].fillna(df["Ranking"].mean())
```

```python
# Assign a new sequential ranking from 1 to the total number of rows in the DataFrame and display unique values.
df["Ranking"] = range(1, len(df)+1)
df["Ranking"].unique()
```

```python
# Display all unique values in the 'Ranking' column after the previous modification.
df["Ranking"].unique()
```

```python
# Display all unique values present in the 'Username' column.
df["Username"].unique()
```

```python
# Find and display the mode (most frequent value) of the 'Username' column.
df["Username"].mode()
```

```python
# Fill any missing (NaN) values in the 'Username' column with the string "Zhong".
df["Username"] = df["Username"].fillna("Zhong")
```

```python
# Display all unique values in the 'Username' column after filling missing values.
df['Username'].unique()
```

```python
# Display all unique values present in the 'Subscribers' column.
df["Subscribers"].unique()
```

```python
# Find and display the mode (most frequent value) of the 'Subscribers' column.
df["Subscribers"].mode()
```

```python
# Fill any missing (NaN) values in the 'Subscribers' column with the string "62.9M".
df["Subscribers"] = df["Subscribers"].fillna("62.9M")
```

```python
# Display all unique values in the 'Subscribers' column after filling missing values.
df["Subscribers"].unique()
```

```python
# Display all unique values present in the 'Uploads' column.
df["Uploads"].unique()
```

```python
# Fill any missing (NaN) values in the 'Uploads' column with 0.
df["Uploads"] = df["Uploads"].fillna(0)
```

```python
# Display all unique values in the 'Uploads' column after filling missing values.
df["Uploads"].unique()
```

```python
# Display all unique values present in the 'Views' column.
df["Views"].unique()
```

```python
# Fill any missing (NaN) values in the 'Views' column with 0.
df["Views"] = df["Views"].fillna(0)
```

```python
# Display all unique values in the 'Views' column after filling missing values.
df["Views"].unique()
```

```python
# Display all unique values present in the 'Country' column.
df["Country"].unique()
```

```python
# Find and display the mode (most frequent value) of the 'Country' column.
df['Country'].mode()
```

```python
# Fill any missing (NaN) values in the 'Country' column with "IN".
df["Country"] = df["Country"].fillna("IN")
```

```python
# Display all unique values in the 'Country' column after filling missing values.
df["Country"].unique()
```

```python
# Display the dimensions (number of rows and columns) of the DataFrame.
df.shape
```

```python
# Re-check for the count of null (missing) values in each column after handling.
df.isnull().sum()
```

**Visualisation**

```python
# Group data by country, sum subscribers, and create a bar plot to visualize distribution.
country_data = df.groupby("Country")["Subscribers"].sum()

plt.figure()
plt.bar(country_data.index, country_data.values)
plt.xlabel("Country")
plt.ylabel("Subscribers")
plt.title("Country-wise subscribers Distribution")
plt.show()
```

```python
# Convert 'Views' to numeric, group by country, sum views, and plot the distribution.
df['Views'] = df['Views'].astype(str).str.replace(",", "")
df['Views'] = df['Views'].astype(float)
country_data = df.groupby("Country")["Views"].sum()

plt.figure()
plt.plot(country_data.index, country_data.values)
plt.xlabel("Country")
plt.ylabel("Total Views")
plt.title("Country-wise views Distribution")
plt.show()
```

```python
# Display the data types of each column in the DataFrame.
df.dtypes
```

```python
# Convert 'Uploads' to numeric, group by country, sum uploads, and plot the distribution.
df['Uploads'] = df['Uploads'].astype(str).str.replace(",", "")
df['Uploads'] = df['Uploads'].astype(float)
uploads = df.groupby("Country")["Uploads"].sum()
uploads = pd.DataFrame(uploads).reset_index()

plt.plot(uploads["Country"], uploads["Uploads"],
         marker="o", mfc="blue", mec="blue")

plt.xlabel("Country")
plt.ylabel("Total Uploads")
plt.title("Uploads by Country")
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of users per country and plot the results.
user_counts = df.groupby("Country")["Username"].count()
user_counts = pd.DataFrame(user_counts).reset_index()

plt.plot(user_counts["Country"], user_counts["Username"],
         marker="o", mfc="blue", mec="blue")

plt.xlabel("Country")
plt.ylabel("Number of Users")
plt.title("Number of Users by Country")
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of rankings per country and plot the results.
user_counts = df.groupby("Country")["Ranking"].count()
user_counts = pd.DataFrame(user_counts).reset_index()

plt.plot(user_counts["Country"], user_counts["Ranking"],
         marker="o", mfc="blue", mec="blue")

plt.xlabel("Country")
plt.ylabel("Ranking")
plt.title("Ranking by Country")
plt.xticks(rotation=90)
plt.show()
```
