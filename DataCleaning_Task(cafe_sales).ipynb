#DATACLEANING


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/dirty_cafe_sales.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

**Handling Transaction id column**

```python
df["Transaction ID"].unique()
```

```python
df["Item"].unique()
```

**Handling Item column**

```python
df['Item'] = df['Item'].replace("UNKNOWN","Juice")
df['Item'].unique()
```

```python
df['Item'].mode()
```

```python
df['Item'] = df['Item'].replace("ERROR","Juice")
df['Item'].unique()
```

**Handling Quantity Column**

```python
df["Quantity"].unique()
```

```python
#Handling Quantity column
df['Quantity'].mode()
```

```python
df['Quantity']=df['Quantity'].replace('ERROR',5)
```

```python
df['Quantity']=df['Quantity'].replace('UNKNOWN',5)
```

```python
df['Quantity'].unique()
```

**Handling Price per unit column**

```python
df['Price Per Unit'].unique()
```

```python
df['Price Per Unit'] = df['Price Per Unit'].replace('ERROR',3)
df['Price Per Unit'] = df['Price Per Unit'].replace('UNKNOWN',3)
df["Price Per Unit"] = df["Price Per Unit"].astype(float)
avg_price = round(df['Price Per Unit'].mean())
avg_price
```

```python
df['Price Per Unit'].unique()
```

**Handling total spent column**

```python
df['Total Spent'].unique()
```

```python
df['Total Spent'] = df['Total Spent'].replace('ERROR',0)
df['Total Spent'] = df['Total Spent'].replace('UNKNOWN',0)
df['Total Spent'] = df['Total Spent'].astype(float)
avg_total = round(df['Total Spent'].mean())
avg_total
```

```python
df['Total Spent'] = df['Total Spent'].replace('ERROR',avg_total)
df['Total Spent'] = df['Total Spent'].replace('UNKNOWN',avg_total)
```

```python
df['Total Spent'].unique()
```

```python
df['Total Spent'] = df['Total Spent'].replace(0.0,avg_total)
```

```python
df['Total Spent'].unique()
```

**Handling Payment Method column**

```python
df['Payment Method'].unique()
```

```python
df['Payment Method'].mode()
```

```python
df['Payment Method'] = df['Payment Method'].replace('UNKNOWN','Digital Wallet')
df['Payment Method'] = df['Payment Method'].replace('ERROR','Digital Wallet')
df['Payment Method'].unique()
```

**handling Location column**

```python
df['Location'].unique()
```

```python
df['Location'].mode()
```

```python
df['Location'] = df['Location'].replace('UNKNOWN','In-store')
df['Location'] = df['Location'].replace('ERROR','Takeaway')
```

```python
df['Location'].unique()
```

**Handling Transaction Date column**

```python
df['Transaction Date'].unique()
```

```python
df['Transaction Date'] = df['Transaction Date'].replace('UNKNOWN','2023-01-01')
df['Transaction Date'] = df['Transaction Date'].replace('ERROR','2023-01-01')
```

```python
df['Transaction Date'].unique()
```
